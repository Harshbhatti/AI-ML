```python
# Random Forest From Scikit-learn on iris dataset:
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# 1.Load the Data

```python
df=datasets.load_iris()
```

```python
df
```

# 2.Get the feature & Labels from the Data

```python
x=df.data
y=df.target
```

```python
x
```

```python
y
```

# 3.Split the data into training & testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

# 4.Standardize the data

```python
sc=StandardScaler()
x_train_std=sc.fit_transform(x_train)
x_test_std=sc.fit_transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

# 5.Chosse Algo & set the Hyperparameters

```python
# after change n_estimators=150,accuracy=100% & misclassified samples=0 & random_state both same
forest=RandomForestClassifier(criterion='entropy',n_estimators=10,random_state=10)
```

# 6.Train the Model

```python
forest.fit(x_train_std,y_train)
```

# 7. Make the prediction using the Model

```python
y_pred=forest.predict(x_test_std)
```

```python
y_pred
```

```python
print('wong predictions out of total:')
print((y_test!=y_pred).sum(),'/',((y_test==y_pred).sum()))
```

```python
print('Percentage Accuracy:',100*accuracy_score(y_test,y_pred))
```
